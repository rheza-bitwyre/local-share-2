```python
# Fiat Assets
# Define parameters
fields = {
    "market": "crypto",        # Options: crypto, equity, stablecoin, fixed_income, commodities, carbon
    "product": "spot",         # Options: spot, futures, options, swaps, spreads
    "country": "GLOBAL"        # ISO 3166-2 Country Code. Only ID or PA are valid
}
# Make GET request
url = "https://api.bitwyre.com/public/pairs"
response = requests.get(url, params=fields)
response_data = response.json()
print(response)
print(response_data)
```

```python
# Ticker
# Define parameters
fields = {
    "instrument": "doge_idr_spot"
}
# Make GET request
url = "https://api.bitwyre.com/public/pairs"
response = requests.get(url, params=fields)
response_data = response.json()
print(response)
print(response_data)
```
