```python
# Spreads
import pandas as pd

df = pd.read_csv('LINA_GTC_spread.csv')

df
```

```python
# Flaging
df['next_spread'] = df['spread'].shift(-1)

df['flag'] = (df['next_spread'] > df['spread']).astype(int)

df.dropna(inplace=True)

df
```

## Features

```python
df_features = df.copy()

# Lag Features
df_features['spread_lag_1'] = df_features['spread'].shift(1)
df_features['spread_lag_2'] = df_features['spread'].shift(2)
df_features['LINA_close_lag_1'] = df_features['LINA_close'].shift(1)
df_features['GTC_close_lag_1'] = df_features['GTC_close'].shift(1)

# Rolling Statistics
df_features['spread_ma_5'] = df_features['spread'].rolling(window=5).mean()
df_features['spread_ma_10'] = df_features['spread'].rolling(window=10).mean()
df_features['spread_std_5'] = df_features['spread'].rolling(window=5).std()
df_features['spread_std_10'] = df_features['spread'].rolling(window=10).std()
df_features['LINA_close_ma_5'] = df_features['LINA_close'].rolling(window=5).mean()
df_features['GTC_close_ma_5'] = df_features['GTC_close'].rolling(window=5).mean()
df_features['LINA_close_std_5'] = df_features['LINA_close'].rolling(window=5).std()
df_features['GTC_close_std_5'] = df_features['GTC_close'].rolling(window=5).std()

# Percentage Changes
df_features['spread_pct_change_1'] = df_features['spread'].pct_change(periods=1)
df_features['spread_pct_change_5'] = df_features['spread'].pct_change(periods=5)
df_features['spread_pct_change_10'] = df_features['spread'].pct_change(periods=10)
df_features['LINA_close_pct_change_1'] = df_features['LINA_close'].pct_change(periods=1)
df_features['GTC_close_pct_change_1'] = df_features['GTC_close'].pct_change(periods=1)

# Price Difference and Ratios
df_features['LINA_GTC_diff'] = df_features['LINA_close'] - df_features['GTC_close']
df_features['LINA_GTC_diff_pct_change'] = df_features['LINA_GTC_diff'].pct_change(periods=1)
df_features['LINA_GTC_ratio'] = df_features['LINA_close'] / df_features['GTC_close']
df_features['LINA_GTC_ratio_pct_change'] = df_features['LINA_GTC_ratio'].pct_change(periods=1)

# Volatility Indicators
df_features['LINA_volatility'] = df_features['LINA_close'].rolling(window=5).std()
df_features['GTC_volatility'] = df_features['GTC_close'].rolling(window=5).std()

# Feature Interactions
df_features['LINA_close_GTC_close_interaction'] = df_features['LINA_close'] * df_features['GTC_close']
df_features['LINA_close_GTC_close_ratio'] = df_features['LINA_close'] / df_features['GTC_close']
df_features['LINA_close_GTC_close_diff'] = df_features['LINA_close'] - df_features['GTC_close']

# Ensure 'timestamp' is in datetime format
df_features['timestamp'] = pd.to_datetime(df_features['timestamp'])

# Now proceed with the feature engineering
df_features['hour'] = df_features['timestamp'].dt.hour
df_features['day_of_week'] = df_features['timestamp'].dt.dayofweek
df_features['month'] = df_features['timestamp'].dt.month

# Flag and Spread Interaction
df_features['spread_flag_interaction'] = df_features['spread'] * df_features['flag']

# Cumulative Features
df_features['LINA_close_cumsum'] = df_features['LINA_close'].cumsum()
df_features['GTC_close_cumsum'] = df_features['GTC_close'].cumsum()

# Rolling Correlation
df_features['LINA_GTC_corr_5'] = df_features['LINA_close'].rolling(window=5).corr(df_features['GTC_close'])
df_features['LINA_GTC_corr_10'] = df_features['LINA_close'].rolling(window=10).corr(df_features['GTC_close'])

df_features.dropna(inplace=True)

df_features
```

# Eval Function

```python
import numpy as np
import polars as pl
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Adjust the evaluate_baseline function to accept only y_pred and y_test
def evaluate_baseline(y_pred, y_test):
    # Convert lists to Numpy arrays
    pred = np.array(y_pred)
    test = np.array(y_test)

    # Evaluation Metrics
    me = np.mean(pred - test)  # Mean Error
    mae = mean_absolute_error(test, pred)  # Mean Absolute Error
    mape = np.mean(np.abs((test - pred) / test)) * 100  # Mean Absolute Percentage Error
    mpe = np.mean((test - pred) / test) * 100  # Mean Percentage Error
    rmse = np.sqrt(mean_squared_error(test, pred))  # Root Mean Squared Error

    r = r2_score(test, pred)  # Coefficient of Determination (R-squared)
    min_max_error = np.abs((np.min(pred) - np.min(test)) + 
                            (np.max(pred) - np.max(test)))  # Min-Max Error

    df = pl.DataFrame({
        f'volatility_prediction': pred,
        f'volatility_actual': test
    })

    df = df.with_columns(
        abs(pl.col(f'volatility_prediction') - pl.col(f'volatility_actual')).alias(f'dif_volatility')
    )

    # overall median abs error
    median_abs_err =  df[f'dif_volatility'].median()

    # overall var abs error
    var_abs_err =  df[f'dif_volatility'].var()

    # Results
    results = {
        "Overall Median Absolute Error" : median_abs_err,
        "Overall Variance Absolute Error" : var_abs_err,
        "Mean Error": me,
        "Mean Absolute Error ": mae,
        "Mean Absolute Percentage Error": mape,
        "Mean Percentage Error": mpe,
        "Root Mean Squared Error": rmse,
        "R-squared": r, 
        "Min-Max Error": min_max_error
    }
    
    return results, df
```

```python
import time
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Separate df into features and target
features_df = df.drop(['timestamp', 'next_spread', 'flag'], axis=1)
target_df = df['next_spread']

# Convert to NumPy arrays
X = features_df.values
y = target_df.values

# Specify parameters for the sliding window approach
num_predictions = 1   # Number of rows to predict
gap = 1               # Gap (number of rows to skip after each window)
max_windows = [1000]    # Maximum number of windows to process
set_limit = True      # Set this to False to process all windows

# Define list of window sizes
window_sizes = [2000]
# window_sizes = list(range(500, 15001, 500))

# List to store results
all_results = []

# List to store resultsAggregate
for max_window in max_windows:
# Loop through each window size
    for window_size in window_sizes:
        print(f'Processing window size: {window_size}')

        # Calculate the number of windows based on dataset size
        num_windows = len(X) - window_size - num_predictions

        # Adjust the starting point if set_limit is True
        start_index = 0
        if set_limit:
            # Only process the latest max_window
            start_index = max(0, len(X) - window_size - num_predictions - max_window)
            num_windows = min(num_windows, max_window)

        # Initialize lists to store actual and predicted values for this window size
        y_vals = []
        y_preds = []

        # Loop through each sliding window with the gap applied
        for window_number in range(start_index, start_index + num_windows, gap):
            
            start = window_number
            end = start + window_size
            X_train = X[start:end]
            y_train = y[start:end]

            X_train_mean = np.mean(X_train, axis=0)
            X_train_std = np.std(X_train, axis=0)
            X_train_normalized = (X_train - X_train_mean) / X_train_std

            # Get the column index for 'cur_avg_vol_4h' from features_df
            close_index = features_df.columns.get_loc('spread')

            # Normalize y_train using the mean and std of cur_avg_vol_4h
            close_mean = X_train[:, close_index].mean()
            close_std = X_train[:, close_index].std()
            y_train_normalized = (y_train - close_mean) / close_std

            # Prepare validation data for prediction
            X_val = X[end:end + num_predictions]
            y_val = y[end:end + num_predictions]

            # Normalize validation data using the statistics from the training set
            X_val_normalized = (X_val - X_train_mean) / X_train_std

            # Track the start time of the window processing
            start_time = time.time()

            # Initialize and fit the model
            model = LinearRegression()
            model.fit(X_train_normalized, y_train_normalized)

            # Predict on validation data
            y_pred_val = model.predict(X_val_normalized)
            # Denormalize y_val and y_pred_val using the mean and std of cur_avg_vol_4h
            y_pred_val_denorm = y_pred_val * close_std + close_mean

            # Track the end time of the window processing
            end_time = time.time()

            # Append actual and predicted values to the lists
            y_vals.append(y_val[0])  # Assuming a single prediction per window
            y_preds.append(y_pred_val_denorm[0])  # Assuming a single prediction per window

        # Now that all windows for this window_size are processed, evaluate the baseline
        eval_results, _ = evaluate_baseline(y_preds, y_vals)

        # Append the window size and evaluation metrics to the results list
        all_results.append({
            'eval_size' : max_window,
            'window_size': window_size,
            **eval_results
        })

# Convert the results to a DataFrame for further analysis
results_df = pd.DataFrame(all_results)

# Print the results DataFrame
results_df
```

```python
import time
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Separate df into features and target
features_df = df_features.drop(['timestamp', 'next_spread', 'flag'], axis=1)
target_df = df_features['flag']

# Convert to NumPy arrays
X = features_df.values
y = target_df.values

# Define list of train-test split proportions
train_test_ratios = [0.95]  # Example proportions for training set

# List to store results
all_results = []

# Loop through each train-test ratio
for train_ratio in train_test_ratios:
    print(f'Processing train-test split ratio: {train_ratio}')

    # Split data chronologically
    split_index = int(len(X) * train_ratio)
    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]

    # Normalize training data
    X_train_mean = np.mean(X_train, axis=0)
    X_train_std = np.std(X_train, axis=0)
    X_train_normalized = (X_train - X_train_mean) / X_train_std

    # Get the column index for 'spread' from features_df
    spread_index = features_df.columns.get_loc('spread')

    # Normalize y_train using the mean and std of the 'spread' column
    spread_mean = X_train[:, spread_index].mean()
    spread_std = X_train[:, spread_index].std()
    y_train_normalized = (y_train - spread_mean) / spread_std

    # Normalize test data using the training set's statistics
    X_test_normalized = (X_test - X_train_mean) / X_train_std

    # Initialize and fit the model
    start_time = time.time()
    model = LinearRegression()
    model.fit(X_train_normalized, y_train_normalized)

    # Predict on the test set
    y_pred_test = model.predict(X_test_normalized)

    # Convert predictions to binary (0 or 1) based on threshold
    y_pred_class = (y_pred_test >= 0).astype(int)

    # Track the end time of the process
    end_time = time.time()
    print(f"Model training and prediction time: {end_time - start_time} seconds")

# Convert results to a DataFrame for further analysis
results_df = pd.DataFrame({
    'y_test': y_test,
    'y_pred_test_classified': y_pred_class
})
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score

# Calculate Accuracy
accuracy = accuracy_score(results_df['y_test'], results_df['y_pred_test_classified'])

# Calculate F1 Score
f1 = f1_score(results_df['y_test'], results_df['y_pred_test_classified'])

# Print results
print(f"Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.2f}")
```
